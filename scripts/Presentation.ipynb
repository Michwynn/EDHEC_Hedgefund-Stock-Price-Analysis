##### Stock price prediction using AutoRegressive Integrated Moving Average (ARMA)

We will perform single time series forecasting using ARMA for each ticker. However, we need to first perform stationarity test using Augmented Dickey-Fuller (ADF) unit root test on log-difference time series of each ticker's price history. The following is the output for each ADF's p-value and whether it rejects the null hypothesis.

```python
adf(raw_prices)
```

Now that stationarity is achieved, we can perform ARMA based on Log-price movements for forecasting. Suppose we want to forecast AAPL's stock, we first need to split the data into training and test sets and then output the forecast values, along with a 50% confidence interval (for a narrower band) to justify predicted values.

```python
draw_ARMA_plot(raw_prices, 'AAPL')
```
